```python
# Setting the terminal postscript with the options
import csv
import os
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math as m
from glob import glob
import seaborn as sns

###### Hybrid files data
foldername='cg/OutputValuesPopulation/'
name=[]
for i in range(0,10):
    for file in os.listdir(foldername):
        if file[:5]=="Out%s."%i:
#             print file
            name.append(file)
for i in range(10,100):
    for file in os.listdir(foldername):
        if file[:6]=="Out%s."%i:
#             print file
            name.append(file)
for file in os.listdir(foldername):
        if file[:7]=="Out%s."%100:
#             print file
            name.append(file)
            
#Population data
Pop={}

#Oxygen data
Pop_ox={}

for file in os.listdir(foldername):
    aux=[]
    if file[:3]=="Out":
        crs = open(foldername+file, "r")
        for columns in ( raw.strip().split() for raw in crs ):
            aux.append(columns[1])
        aux=np.array(map(float,aux))
        Pop[file]=aux
foldername='cg/OutputValuesOxygen/'
for file in os.listdir(foldername):
    aux=[]
    if file[:3]=="Out":
        crs = open(foldername+file, "r")
        for columns in ( raw.strip().split() for raw in crs ):
            aux.append(columns[1])
        aux=np.array(map(float,aux))
        Pop_ox[file]=aux
        
###### Coarse grained files data
foldername='cgrk/OutputValuesPopulation/'
            
#Population data
Pop_cg={}

#Oxygen data
Pop_ox_cg={}

for file in os.listdir(foldername):
    aux=[]
    if file[:3]=="Out":
        crs = open(foldername+file, "r")
        for columns in ( raw.strip().split() for raw in crs ):
            aux.append(columns[1])
        aux=np.array(map(float,aux))
        Pop_cg[file]=aux
foldername='cgrk/OutputValuesOxygen/'
for file in os.listdir(foldername):
    aux=[]
    if file[:3]=="Out":
        crs = open(foldername+file, "r")
        for columns in ( raw.strip().split() for raw in crs ):
            aux.append(columns[1])
        aux=np.array(map(float,aux))
        Pop_ox_cg[file]=aux
```
