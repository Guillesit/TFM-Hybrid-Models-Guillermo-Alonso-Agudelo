# Scrip to compare, calulate errors, ect. between a population with age structure and a modification to simplify it. 

Charge the files.

```python
# Setting the terminal postscript with the options
import csv
import os
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math as m
from glob import glob
import seaborn as sns



foldername='Pruebas_Pilar_UCL/Ceros2/OutputValuesPopulation/'
name=[]
for i in range(0,10):
    for file in os.listdir(foldername):
        if file[:6]=="Out00%s"%i:
            name.append( file)
for i in range(10,200):
    for file in os.listdir(foldername):
        if file[:6]=="Out0%s"%i:
            name.append( file)
for i in range(100,101):
    for file in os.listdir(foldername):
        if file[:6]=="Out%s"%i:
            name.append( file)
            
#Population data
Pop2={}
Pop_mod2={}
Pop3={}
Pop_mod3={}
Pop4={}
Pop_mod4={}
Pop5={}
Pop_mod5={}
Pop6={}
Pop_mod6={}
Pop_stoc6={}
list_age=[Pop2,Pop3,Pop4,Pop5,Pop6]
list_mod=[Pop_mod2,Pop_mod3,Pop_mod4,Pop_mod5,Pop_mod6]

#Oxygen data
Pop2_ox={}
Pop_mod2_ox={}
Pop3_ox={}
Pop_mod3_ox={}
Pop4_ox={}
Pop_mod4_ox={}
Pop5_ox={}
Pop_mod5_ox={}
Pop6_ox={}
Pop_mod6_ox={}
list_age_ox=[Pop2_ox,Pop3_ox,Pop4_ox,Pop5_ox,Pop6_ox]
list_mod_ox=[Pop_mod2_ox,Pop_mod3_ox,Pop_mod4_ox,Pop_mod5_ox,Pop_mod6_ox]
        
        
for j in range(2,7):
    foldername='Pruebas_Pilar_UCL/Ceros%s/OutputValuesPopulation/'%j
    list_age[j-2]={}
    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_age[j-2][file]=aux
        
    foldername='Tests100ParaleloJuanMarzo/MetodoSimplificadoCeros%s/OutputValuesPopulation/'%j
    list_mod[j-2]={}

    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            #print file
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_mod[j-2][file]=aux
            
    foldername='Pruebas_Pilar_UCL/Ceros%s/OutputValuesOxygen/'%j
    list_age_ox[j-2]={}
    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_age_ox[j-2][file]=aux
        
    foldername='Tests100ParaleloJuanMarzo/MetodoSimplificadoCeros%s/OutputValuesOxygen/'%j
    list_mod_ox[j-2]={}

    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_mod_ox[j-2][file]=aux
        
Pop2=list_age[0]
Pop_mod2=list_mod[0]
Pop3=list_age[1]
Pop_mod3=list_mod[1]
Pop4=list_age[2]
Pop_mod4=list_mod[2]
Pop5=list_age[3]
Pop_mod5=list_mod[3]
Pop6=list_age[3]
Pop_mod6=list_mod[3]
```

## Error calculation

```python
error2=[]
for l in name:
    aux=np.power(Pop2[l]-Pop_mod2[l],2)/101
    pop=np.sum(aux)
#     print m.sqrt(100.0)
    error2.append(m.sqrt((pop)/m.sqrt(len(Pop2[l]))))
    
error3=[]
for l in name:
    aux=np.power(Pop3[l]-Pop_mod3[l],2)/101
    pop=np.sum(aux)
    error3.append(m.sqrt((pop)/m.sqrt(len(Pop3[l]))))
error4=[]
for l in name:
#     print l
    aux=np.power(Pop4[l]-Pop_mod4[l],2)/101
    pop=np.sum(aux)
    error4.append(m.sqrt((pop)/m.sqrt(len(Pop4[l]))))
#     print sum (Pop3[l])

error5=[]
for l in name:
    aux=np.power(Pop5[l]-Pop_mod5[l],2)/101
    pop=np.sum(aux)
    error5.append(m.sqrt((pop)/m.sqrt(len(Pop5[l]))))
error6=[]
for l in name:
    aux=np.power(Pop6[l]-Pop_mod6[l],2)/101
    pop=np.sum(aux)
    error6.append(m.sqrt((pop)/m.sqrt(len(Pop6[l]))))
```

```python
###ERROR (mean(Pop)-mean(Pop_mod))/((sed+sed)/2)
#### Sale muy grande!!! si le quito el cuadrado sale algo razonable


error2=[]
for l in name:
    esd=np.std(Pop2[l])/m.sqrt(len(Pop2[l]))#Standar Error of the mean = std/sqrt(N)
    esd2=np.std(Pop_mod2[l])/m.sqrt(len(Pop_mod2[l]))# Standar Error of the mean = std/sqrt(N)
    pop=(np.mean(Pop2[l])-np.mean(Pop_mod2[l]))/(esd+esd2)#standard error of the difference between independent means
    error2.append(abs(pop))
# plt.plot(error2,'o-',linewidth=1.,label='Ceros2')
error3=[]
for l in name:
    esd=np.std(Pop3[l])/m.sqrt(len(Pop3[l]))#Standar Error of the mean = std/sqrt(N)
    esd2=np.std(Pop_mod3[l])/m.sqrt(len(Pop_mod3[l]))# Standar Error of the mean = std/sqrt(N)
    pop=(np.mean(Pop3[l])-np.mean(Pop_mod3[l]))/(esd+esd2)#standard error of the difference between independent means
    error3.append(abs(pop))
# plt.plot(error3,'o-',linewidth=1.,label='Ceros3')
error4=[]
for l in name:
    esd=np.std(Pop4[l])/m.sqrt(len(Pop4[l]))#Standar Error of the mean = std/sqrt(N)
    esd2=np.std(Pop_mod4[l])/m.sqrt(len(Pop_mod4[l]))# Standar Error of the mean = std/sqrt(N)
    pop=(np.mean(Pop4[l])-np.mean(Pop_mod4[l]))/(esd+esd2)#standard error of the difference between independent means
    error4.append(abs(pop))
plt.plot(error4,'o-',linewidth=1.,label='Ceros4')
error5=[]
for l in name:
    esd=np.std(Pop5[l])/m.sqrt(len(Pop5[l]))#Standar Error of the mean = std/sqrt(N)
    esd2=np.std(Pop_mod5[l])/m.sqrt(len(Pop_mod5[l]))# Standar Error of the mean = std/sqrt(N)
    pop=(np.mean(Pop5[l])-np.mean(Pop_mod5[l]))/(esd+esd2)#standard error of the difference between independent means
    error5.append(abs(pop))
plt.plot(error5,'o-',linewidth=1.,label='Ceros5')
error6=[]
for l in name:
    esd=np.std(Pop6[l])/m.sqrt(len(Pop6[l]))#Standar Error of the mean = std/sqrt(N)
    esd2=np.std(Pop_mod6[l])/m.sqrt(len(Pop_mod6[l]))# Standar Error of the mean = std/sqrt(N)
    pop=np.mean(Pop6[l])-np.mean(Pop_mod6[l])#standard error of the difference between independent means
    error6.append(abs(pop)/((esd+esd2)))

plt.plot(error6,'o-',linewidth=1.,label='Ceros6')

plt.legend(loc=2)
# plt.xlim([3.5,8.5])
plt.grid(True)
# plt.xticks(np.arange(0,100,100/6),np.arange(0,5000,50000/10), fontsize = 15) # work on current fig
plt.yticks( fontsize = 15) # work on current fig
plt.xticks(range(0,110,100/5),range(0,550000,500000/5), fontsize = 15) # work on current fig
plt.yticks( fontsize = 15) # work on current fig

plt.xlabel('time', fontsize=15)
plt.ylabel(r'$\frac{abs(\bar{Pop_{age}}-\bar{Pop_{mod}})}{(SED_{age}+SED_{mod})}$', fontsize=15)
#plt.savefig('error4.eps', dpi=100, bbox_inches='tight')
```

## Error recomendado por Tomas $\frac{\sqrt{(\sum_x P_{age}-P_{mod})^2}}{\sqrt{\sum_x P_{age}^2}}$

```python
# Error recomendado por Tomas
# sqrt(Suma_x ('estructura de edad'(x_i)-'modificado'(x_i))^2)
#     / sqrt(Suma_x 'estructura de edad'(x_i)^2).


error2=[]
for l in name:
    aux=np.power(Pop2[l]-Pop_mod2[l],2)
    pop=np.sum(aux)/sum(np.power(Pop2[l],2))
    error2.append(np.sqrt((pop)))
    
error3=[]
for l in name:
    aux=np.power(Pop3[l]-Pop_mod3[l],2)
    pop=np.sum(aux)/sum(np.power(Pop3[l],2))
    error3.append(np.sqrt((pop)))
error4=[]
for l in name:
#     print l
    aux=np.power(Pop4[l]-Pop_mod4[l],2)
    pop=np.sum(aux)/sum(np.power(Pop4[l],2))
    error4.append(np.sqrt((pop)))
#     print sum (Pop3[l])

error5=[]
for l in name:
    aux=np.power(Pop5[l]-Pop_mod5[l],2)
    pop=np.sum(aux)/sum(np.power(Pop5[l],2))
    error5.append(np.sqrt((pop)))
error6=[]
for l in name:
    aux=np.power(Pop6[l]-Pop_mod6[l],2)
    pop=np.sum(aux)/sum(np.power(Pop6[l],2))
    error6.append(np.sqrt((pop)))
plt.plot(error2,'o-',linewidth=1.,label='$O$(-3)')
plt.plot(error3,'o-',linewidth=1.,label='$O$(-4)')
plt.plot(error4,'o-',linewidth=1.,label='$O$(-5)')
plt.plot(error5,'o-',linewidth=1.,label='$O$(-6)')
plt.plot(error6,'o-',linewidth=1.,label='$O$(-7)')
plt.legend(loc=1)
# plt.xlim([3.5,8.5])
plt.grid(True)
# y=[0, ]
plt.xticks(range(0,110,100/5),range(0,550000,500000/5), fontsize = 15) # work on current fig
plt.yticks( fontsize = 15) # work on current fig

plt.xlabel('time', fontsize=15)
plt.ylabel(r'$\frac{\sqrt{\sum_x (P_{age}-P_{mod})^2}}{\sqrt{\sum_x P_{age}^2}}$', fontsize=15)
plt.savefig('error_Tomas.eps', dpi=100, bbox_inches='tight')
```

## Calulation of the wave front differnce (Error TW)

```python
error=[]
i=0
for m in list_age:
#     print i
    plt.figure()
    m2=list_mod[i]
#     print m, m2
    error=[]
    for l in name:
        Pop_aux=np.array(m[l])
        Pop_maux=np.array(m2[l])
        if len(Pop_aux[Pop_aux==0])>1:
            aux=Pop_aux[np.argmax(Pop_aux==0)-1]-Pop_maux[np.argmax(Pop_maux==0)-1]
        else:
            aux=abs(Pop_aux[-1]-Pop_maux[-1])
        error.append(abs(aux))
    i=i+1
    plt.plot(error,'-',linewidth=3.,label='$O$(-%s)'%(i+2))


    plt.title(r'Wave front error between $P_{age}$ and $P_{mod}$ $O$(-%s)'%(i+2))
    #plt.legend(loc=1,frameon=True, fancybox=True, shadow=True)
    # plt.xlim([3.5,8.5])
    plt.grid(True)
    # y=[0, ]
    plt.xticks(range(0,110,100/5),range(0,550000,500000/5), fontsize = 15) # work on current fig
    plt.yticks( fontsize = 15) # work on current fig
    plt.xticks( fontsize = 15)
    plt.xlabel('Time(t)', fontsize=15)
    plt.ylabel(r'$m^2$', fontsize=15)
    plt.savefig('error_TW_%s.eps'%(i+1), dpi=100, bbox_inches='tight')
```
