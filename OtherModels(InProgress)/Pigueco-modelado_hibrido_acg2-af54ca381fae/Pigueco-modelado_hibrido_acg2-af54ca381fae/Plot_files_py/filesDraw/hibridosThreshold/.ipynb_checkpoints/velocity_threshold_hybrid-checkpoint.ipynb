```python
# Setting the terminal postscript with the options
import csv
import os
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math as m
from glob import glob
import seaborn as sns

###### Hybrid files data
foldername='OutputValuesPopulation2000/'
name=[]
for i in range(0,10):
    for file in os.listdir(foldername):
        if file[:5]=="Out%s."%i:
#             print file
            name.append(file)
for i in range(10,100):
    for file in os.listdir(foldername):
        if file[:6]=="Out%s."%i:
#             print file
            name.append(file)
for file in os.listdir(foldername):
        if file[:7]=="Out%s."%100:
#             print file
            name.append(file)
            
#Population data
Pop0={}
Pop1={}
Pop2={}
Pop3={}
Pop4={}
Pop5={}
Pop6={}
Pop7={}
Pop8={}

list_pop=[Pop0,Pop1,Pop2,Pop3,Pop4,Pop5,Pop6,Pop7,Pop8]

#Oxygen data
Pop0_ox={}
Pop1_ox={}
Pop2_ox={}
Pop3_ox={}
Pop4_ox={}
Pop5_ox={}
Pop6_ox={}
Pop7_ox={}
Pop8_ox={}

list_ox=[Pop0_ox,Pop1_ox, Pop2_ox,Pop3_ox,Pop4_ox,Pop5_ox,Pop6_ox,Pop7_ox,Pop8_ox]




threshold_values=[10,100,500, 1000,  1500, 2000,2500,3000,3500]
T=0
for threshold in threshold_values:
    foldername='OutputValuesPopulation%s/'%threshold
    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_pop[T][file]=aux
    foldername='OutputValuesOxygen%s/'%threshold
    for file in os.listdir(foldername):
        aux=[]
        if file[:3]=="Out":
            crs = open(foldername+file, "r")
            for columns in ( raw.strip().split() for raw in crs ):
                aux.append(columns[1])
            aux=np.array(map(float,aux))
            list_ox[T][file]=aux
    T=T+1
```

```python
#Same time snapshot of Population
nu = 25000 #aprox 1/deathrate
i=0
num_plots =10
for l in name[0::5]:
    plt.figure()
    sns.set_style("whitegrid")
    num_plots = 42
    colormap =sns.cubehelix_palette( start=1,rot=1, as_cmap=True)#plt.cm.plasma#gist_ncar
    plt.gca().set_color_cycle([colormap(j) for j in np.linspace(0, .9, num_plots)])
    i=0
    for threshold in threshold_values:
        Pop_haux=np.array(list_pop[i][l])
        plt.plot(Pop_haux,'-',linewidth=3, label="%s"%threshold)
        i=i+1
    plt.title('$O$(-%s)'%(i+2))
    plt.legend()
    plt.yticks( fontsize = 15) # work on current fig
    plt.xticks( fontsize = 15)
    plt.xlabel('space (X)', fontsize=15)
    plt.ylabel(r'Population', fontsize=15)
    plt.savefig('Compare_pop%s.eps'%(7), dpi=100, bbox_inches='tight')
```

# Velocity as function threshold (hybrid)

```python
plt.figure()
nu = 25000 #aprox 1/deathrate
i=0
for threshold in threshold_values:
    Pop_maux_mean=[]
    for l in name[0::]:
        Pop_maux=np.array(list_pop[i][l])
        Pop_maux_mean.append(np.mean(np.where((Pop_maux<Pop_maux[0]) & (Pop_maux>0.01))[0]))
    i=i+1
    plt.plot(Pop_maux_mean,'-',linewidth=3,label="%s"%threshold)

plt.title('$O$(-%s)'%(7))
# plt.legend([pop_mod, pop_sto], ['$P_{mod}$', '$P_{stoc}$'],frameon=True, fancybox=True, shadow=True,fontsize=15,loc=0)

# plt.xlim(50,100)
plt.ylim(0,150)
plt.xticks(range(0,100,100/5),range(0,5500000/nu,5000000/(nu*5)), fontsize = 15) # work on current fig
plt.legend()
plt.yticks( fontsize = 15) # work on current fig
plt.xticks( fontsize = 15)
plt.xlabel('time', fontsize=15)
plt.ylabel(r'Mean X axis TW pop', fontsize=15)
plt.savefig('mean_position_TW%s.eps'%7, dpi=100, bbox_inches='tight')
    
```

```python
plt.figure()
nu = 25000 #aprox 1/deathrate
i=0
v_mean = []
t_total = 5000000
for threshold in threshold_values:
    Pop_maux_mean=[]
    for l in name[10::]:
        Pop_maux=np.array(list_pop[i][l])
        Pop_maux_mean.append(np.mean(np.where((Pop_maux<Pop_maux[0]) & (Pop_maux>0.01))[0]))
    i=i+1
    v_mean.append((Pop_maux_mean[-1]-Pop_maux_mean[0])/t_total)
plt.plot(v_mean,'o' ,linewidth=3)
plt.title('$O$(-%s)'%(7))
# plt.legend([pop_mod, pop_sto], ['$P_{mod}$', '$P_{stoc}$'],frameon=True, fancybox=True, shadow=True,fontsize=15,loc=0)

plt.xlim(-0.1,8.1)
# plt.ylim(0,150)
plt.xticks(range(0,9),threshold_values, fontsize = 15, rotation=-40 ) # work on current fig
plt.ticklabel_format(style='sci', axis='y',scilimits=(0,0))
plt.legend()
plt.yticks( fontsize = 15) # work on current fig
plt.xticks( fontsize = 15)
plt.xlabel('Threshold', fontsize=15)
plt.ylabel(r'$\bar{v}$', fontsize=15)
plt.savefig('mean_position_TW%s.eps'%7, dpi=100, bbox_inches='tight')
    
```

```python
plt.figure()



t_step = 50000
nu=25000 #1/deathrate (en verdad es 24000)
aux=10
T= 0
for threshold in threshold_values:
    vel_h = []
    Pop_maux_mean=[]
    for i in range(0,len(name)-aux):
        l=name[i]
        m=name[i+aux] 

        Pop_haux=np.array(list_pop[T][l])
        Pop_haux_mean = (np.mean(np.where((Pop_haux<Pop_haux[0]) & (Pop_haux>0.01))[0]))
        
        Pop_haux=np.array(list_pop[T][m])
        Pop_haux_mean_next = (np.mean(np.where((Pop_haux<Pop_haux[0]) & (Pop_haux>0.01))[0]))

        vel_h.append(np.array(Pop_haux_mean_next-Pop_haux_mean)/(aux*t_step))


    plt.plot(vel_h,'o--',linewidth=3,label="%s"%threshold)
    T = T+1


plt.title('$O$(-%s)'%(7))
# plt.legend([pop_mod, pop_sto], ['$P_{mod}$', '$P_{stoc}$'],frameon=True, fancybox=True, shadow=True,fontsize=15,loc=0)

# plt.xlim(10,100)
# plt.ylim(0,150)
plt.xticks(range(0,100,100/5),range(0,5500000/nu,5000000/(nu*5)), fontsize = 15) # work on current fig
plt.legend()
plt.yticks( fontsize = 15) # work on current fig
plt.xticks( fontsize = 15)
plt.xlabel('time', fontsize=15)
plt.ylabel(r'velocity pop', fontsize=15)
plt.savefig('mean_position_TW_h_cg.eps', dpi=100, bbox_inches='tight')
```
